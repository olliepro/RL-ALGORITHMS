```python
import json
import matplotlib.pyplot as plt
import numpy as np

with open("../../results_3.json") as f:
    data = json.load(f)

y = np.mean(data["iEF"], axis=0)
x = np.arange(1, len(y) + 1)
y_err = np.std(data["iEF"], axis=0)

plt.fill_between(x, y - y_err, y + y_err, alpha=0.3)
plt.plot(x, y, label="iEF")

y = np.mean(data["EF"], axis=0)
y_err = np.std(data["EF"], axis=0)

plt.fill_between(x, y - y_err, y + y_err, alpha=0.3)
plt.plot(x, y, label="EF")

plt.xlabel("Number of Policy Iterations")
plt.ylabel("Reward")
plt.legend()
plt.show()
```
